```python
# Program To Read video
# and Extract Frames
import cv2
  
# Function to extract frames
def FrameCapture(path):
      
    # Path to video file
    vidObj = cv2.VideoCapture(path)
  
    # Used as counter variable
    count = 0
  
    # checks whether frames were extracted
    success = 1
  
    while success:
  
        # vidObj object calls read
        # function extract frames
        success, image = vidObj.read()
  
        # Saves the frames with frame-count
        try : 
          cv2.imwrite("frame%d.jpg" % count, image)
        except :
          pass 
        count += 1
  
# Driver Code
```

```python
model_type='yolov5x'
model = torch.hub.load('ultralytics/yolov5',model_type , pretrained=True)
```

```python
dir = '/content/gdrive/My Drive/Projects /RL_Driving/Front_view/'
imgs = [dir +'tk.jpg']  # batched list of images
rand_img = imread(imgs[0])
img=rand_img
# Inference
results = model(imgs)
class_labels=results.names
rect_tensor=results.xyxy[0]
for rect in rect_tensor:
  x1=int(rect[0])
  y1=int(rect[1])
  x2=int(rect[2])
  y2=int(rect[3])
  class_id=int(rect[5])
  text=class_labels[class_id]
  #print(x1,x2,y1,y2)

  img=cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,0),2)
  #text="class name"
  xav=int((x1+x2)/2)
  yav=int((y1+y2)/2)
  position=(xav,yav)
  img=add_text_label(text,position,img)
plt.imshow(img)
```
